```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_pretrain_entities = np.load("num_pretrain_entities/triviaqa_validation.npy", )
plt.hist(num_pretrain_entities, bins=10, range=(0, 1e7))

plt.yscale('log')#, nonposy='clip')
plt.title("TriviaQA")
plt.xlabel("Num Mentions in Pretrain Corpus")
plt.ylabel("Frequency")
```

```python
num_pretrain_entities = np.load("num_pretrain_entities/wikiplots_test_points.npy", )
plt.hist(num_pretrain_entities, bins=10, range=(0, 2000))

plt.yscale('log')#, nonposy='clip')
plt.title("Plot Generation")
plt.xlabel("Num Mentions in Pretrain Corpus")
plt.ylabel("Frequency")

plt.ylim(0, 1000)
```
