```python
import numpy as np 
import matplotlib.pyplot as plt
import os

languages = ["en", "de", "fr", "it", "vi", "ar", "ro", "sk", "ca", "hr", "bn", "ne", "mr", "kn"]


def get_accuracies_all(run_names, checkpoints):
    accuracies_all = []
    for i in range(len(run_names)):
        run_accuracies = []
        for language in languages:
            run_name = run_names[i]
            checkpoint = checkpoints[i]
            language_correct_all = np.load(f"ckpts/{run_name}/checkpoint_{checkpoint}/hf_model/{language}_correct_all.npy")
            run_accuracies.append(language_correct_all.mean())
        accuracies_all.append(run_accuracies)
    return np.array(accuracies_all)
```
